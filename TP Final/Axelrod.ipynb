```python
import matplotlib.pyplot as plt
from scipy.optimize import minimize
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
import random
import time
import pandas as pd
import math
import csv
import os

t0=time.time()
###########################################################
###########################################################
# Acordate que en casa los ldata tienen que empezar con Tests/
# En la facultad va directo el nombre del archivo
############################################################
############################################################

def scan(cant,lista):
    i=0
    for x in lista:
        print(x)
        i+=1
        if i>cant:
            break
            
def Tiempo():
    t1=time.time()
    print("Esto tardó {} segundos".format(t1-t0))


def ldata(archive):
        f = open(archive)
        data = []
        for line in f:
            col = line.split("\t")
            col = [x.strip() for x in col]
            data.append(col)
        return data 
    
# Vamos a leer los archivos con esta celda
#--------------------------------------------------------------------------

CarpCheck=[[root,files] for root,dirs,files in os.walk(".")]

# Los primero dos elementos de CarpCheck no les des bola, porque son
# la mismísma carpeta en la que estás y porque el segundo es ipynb checkpoints
# Eso es algo del notebook, no del TP.

Puntos = []

for x in CarpCheck:
#     dada = x[0].split("\\")
    Puntos.append(x[0])

# Con esto tengo los nombres de todas las carpetas que estoy usando
# Fijate que si cambias las carpetas esto cambia

CarpCheck=[[root,files] for root,dirs,files in os.walk(Puntos[2])]
Casbas = CarpCheck[0][1]

# Con esto tengo los nombres de todos los archivos en todas las carpetas
# Casbas tiene los nombres de los archivos del caso básico del modelo de Axelrod

#-------------------------------------------------------------------------

# Voy a levantar la red y de ahí graficarla usando el pcolormesh
# ¿Qué forma tienen mis datos? Mis datos son matrices con N filas y N*F columnas
# ¿Qué forma tienen los títulos de mis archivos? Son "Red_F=$_Q=$"
# ¿Cómo grafico el pcolormesh? Hasta donde entiendo, lo que voy a tener que hacer es asignarle
# un número a cada individuo en la red.
# ¿Cómo asigno los números? Para eso voy a armar un diccionario que asigne número a cada tipo de
# vector. Necesito crear un diccionario cuyos índices sean los vectores, y sus valores sean números


# Partamos de leer el archivo con la info

Datos = ldata("./Caso Basico/{}".format(Casbas[0]))
Datos = [fila[0:len(Datos[0])-1] for fila in Datos] # Este paso es para eliminar ese vacío final en
# las filas

# A partir del nombre del archivo, obtengo el valor de F y Q
# Repito, tienen el formato "Red_F=$_Q=$.txt"

dada = Casbas[0].split("_")
dede = dada[1].split("=")
F = int(dede[1]) # Acá saco el valor de F

dede = dada[2].split("=")
dodo = dede[1].split(".")
Q = int(dodo[0]) # Acá saco el valor de Q

# Empecemos creando el diccionario

Numeros = dict()

# Defino mi Vector, que va a ser el índice de mi diccionario. Tiene que tener tamaño F

Vector = np.zeros(F)

# Con esto estoy recorriendo los primeros Q elementos
# El motivo de que haya un -1 en la primer coordenada, es para que el primer elemento
# Sea efectivamente el elemento nulo
Vector[0]-=1
n = 0

for i in range(Q):
    Vector[0]+=1
    Numeros[Vector] = n
    n+=1
Vector[0] = -1

print(Numeros)




Tiempo()
```
